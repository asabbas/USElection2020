```python
import json
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1.inset_locator import inset_axes
```

```python
with open('nyt-2020-election-scraper/results.json') as f:
    data = json.load(f)
```

```python
# General Inputs
stateName = data["data"]["races"][38]["state_name"]
stateElecVote = data["data"]["races"][38]["electoral_votes"]
stateTotCountVote = data["data"]["races"][38]["votes"]
# Figure 1
timeArrayLength = len(data["data"]["races"][38]["timeseries"])
time = range(0,timeArrayLength)
# print(timeArrayLength)
bidenCountTimeSeries = []
trumpCountTimeSeries = []
timeLabelTimeSeries = []
netTotalVote = []
prevTotalVote = 0

marginTrump = []
marginBiden = []
# bidenCount = 0
# trumpCount = 0

# totalVote = float(data["data"]["races"][3]["timeseries"][1]["votes"])
# bidenVoteShare = float(data["data"]["races"][3]["timeseries"][1]["vote_shares"]["bidenj"])
# trumpVoteShare = float(data["data"]["races"][3]["timeseries"][1]["vote_shares"]["trumpd"])
# print(trumpVoteShare,bidenVoteShare)
# prevBidenCount = 0
# prevTrumpCount = 0
for i in range(0,timeArrayLength):
    
    currTotalVote = float(data["data"]["races"][38]["timeseries"][i]["votes"])
    bidenVoteShare = float(data["data"]["races"][38]["timeseries"][i]["vote_shares"]["bidenj"])
    trumpVoteShare = float(data["data"]["races"][38]["timeseries"][i]["vote_shares"]["trumpd"])
    timeLabel =  data["data"]["races"][38]["timeseries"][i]["timestamp"]
    
    #for figure 1
    bidenCount = bidenVoteShare * currTotalVote
    trumpCount = trumpVoteShare * currTotalVote
    
    bidenCountTimeSeries.append(bidenCount)
    trumpCountTimeSeries.append(trumpCount)
    timeLabelTimeSeries.append(timeLabel)
    netTotalVote.append(currTotalVote-prevTotalVote)
    prevTotalVote = currTotalVote
    
    
    #for figure 2
    
    marginTrump.append(trumpCount - bidenCount) 
    marginBiden.append(bidenCount - trumpCount)
#     bidenCount = bidenVoteShare * currTotalVote - prevBidenCount
#     bidenCountSum = bidenCount + prevBidenCount
#     prevBidenCount = bidenCount
# #     print(bidenCount)

#     trumpCount = trumpVoteShare * currTotalVote - prevTrumpCount
#     trumpCountSum = trumpCount + prevTrumpCount
#     prevTrumpCount = trumpCount
    
    
#     if i == 0:
#         x = float(0.00)
#     else:
#         x = float(data["data"]["races"][38]["timeseries"][i]["votes"]) - float(data["data"]["races"][38]["timeseries"][i-1]["votes"])
    
#     netTotalVote.append(x)
    
#     bidenCount = bidenVoteShare*x + bidenCount
#     trumpCount = trumpVoteShare*x + trumpCount
#     bidenCountTimeSeries.append(bidenCountSum)
#     trumpCountTimeSeries.append(trumpCountSum)

plt.style.use('fivethirtyeight')
fig, ax = plt.subplots(2,2, figsize=([45,30]))

ax[0,0].plot(time,bidenCountTimeSeries, 'b',lw=12,label= "Biden")
ax[0,0].plot(time,trumpCountTimeSeries, 'r',lw=12,label="Trump")
ax[0,0].set_title(stateName +' Total Votes with ' + str(stateElecVote)+ ' Electoral Votes', fontsize = 38)
ax[0,0].set_ylabel("Vote Counts",fontsize = 38)
ax[0,0].set_xlabel("Vote Release Overtime", fontsize = 38)
ax[0,0].tick_params(axis='both', which='major', labelsize=34)
tx = ax[0,0].yaxis.get_offset_text()
tx.set_fontsize(34)
ax[0,0].legend(fontsize = 28)

axinset1 = inset_axes(ax[0,0], width="50%", height="40%", loc=4)

axinset1.plot(time,bidenCountTimeSeries, 'b', lw=8,label= "Biden")
axinset1.plot(time,trumpCountTimeSeries,'r', lw=8,label="Trump")
axinset1.plot(time, netTotalVote, lw=8,label="Net Votes Added")
axinset1.legend(loc = 'best')
axinset1.tick_params(labelleft=False, labelbottom=False)
axinset1.grid(False)


# axinset2 = inset_axes(ax[0,0], width="30%", height="30%")
# axinset2.plot(time[timeArrayLength-50 : timeArrayLength],bidenCountTimeSeries[timeArrayLength-50 : timeArrayLength], 'b',label= "Biden")
# axinset2.plot(time[timeArrayLength-50 : timeArrayLength],trumpCountTimeSeries[timeArrayLength-50 : timeArrayLength],'r',label="Trump")
# axinset2.grid(False)


# Figure 2

ax[0,1].plot(time,marginBiden, 'b',lw=12,label= "Biden")
ax[0,1].plot(time,marginTrump, 'r',lw=12,label= "Trump")
ax[0,1].set_title('Total Differential Votes Overtime', fontsize = 38)
ax[0,1].set_ylabel(r"$\Delta$Vote Counts",fontsize = 38)
ax[0,1].set_xlabel("Vote Release Overtime", fontsize = 38)
ax[0,1].tick_params(axis='both', which='major', labelsize=34)
ax[0,1].legend(fontsize = 28)


# Figure 3

if (data["data"]["races"][38]["candidates"][0]["candidate_key"] == "bidenj"):
    bidenVotePerc = round((100*float(data["data"]["races"][38]["candidates"][0]["votes"])/float(stateTotCountVote)),4)
    trumpVotePerc = round((100*float(data["data"]["races"][38]["candidates"][1]["votes"])/float(stateTotCountVote)),4)
else:
    bidenVotePerc = round((100*float(data["data"]["races"][38]["candidates"][1]["votes"])/float(stateTotCountVote)),4)
    trumpVotePerc = round((100*float(data["data"]["races"][38]["candidates"][0]["votes"])/float(stateTotCountVote)),4)

labels = ['Biden', "Trump", "Others"]
colors = ['b','r','g']
sizes = [bidenVotePerc*100, trumpVotePerc*100, 100-abs(trumpVotePerc - bidenVotePerc)]
wedges, texts, autotexts = ax[1,0].pie(sizes, colors = colors, autopct='%1.2f%%',shadow=True, startangle=0, textprops = {"color":"w","fontweight":"bold","fontsize":35})
ax[1,0].set_title("2020 Candidates' Vote Percentage for " + stateName, fontsize = 38)
ax[1,0].legend(wedges, labels, fontsize =34,loc="best")
ax[1,0].axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.


# Figure 4

# year = ['2016', '2020']
# percDEM = [60, bidenVotePerc]
# c = [40,trumpVotePerc]
# percOther = [40, 1.5]
# ind = [0,1]
# ax[1,1].bar(ind,percDEM , width=0.8, label='DEM', color='b', bottom=percOther + percGOP)
# ax[1,1].bar(ind, percOther, width=0.8, label='Others', color='g')
# ax[1,1].bar(ind, percOther, width=0.8, label='Others', color='c')





years = ['2016', '2020']
vote2016 = float(data["data"]["races"][38]["votes2016"])
dem2016 = float(data["data"]["races"][38]["clinton2016"])
gop2016 = float(data["data"]["races"][38]["trump2016"])
dem2016VotePerc = 100 * (dem2016 / vote2016)
gop2016VotePerc = 100* (gop2016 / vote2016)
# print(gop2016VotePerc + dem2016VotePerc)

percDEM = np.array([dem2016VotePerc,bidenVotePerc])
percOthers = np.array([100 - abs(gop2016VotePerc + dem2016VotePerc), 100 - abs(trumpVotePerc + bidenVotePerc)])
percGOP = np.array([gop2016VotePerc,trumpVotePerc])
ind = [x for x, _ in enumerate(years)]

bar1= ax[1,1].bar(ind, percDEM, width=0.4, label='DEM', color='b', bottom=percOthers+percGOP)
bar2= ax[1,1].bar(ind, percGOP, width=0.4, label='GOP', color='r', bottom=percOthers)
bar3= ax[1,1].bar(ind, percOthers, width=0.4, label='Others', color='g')


ax[1,1].set_xticks(ind)
ax[1,1].set_xticklabels(years, rotation=0)
ax[1,1].set_title('2016 Vs 2020 for ' + stateName, fontsize = 38)
ax[1,1].set_ylabel("Vote Percent [%]", fontsize = 38)
ax[1,1].set_xlabel("Years", fontsize = 38)
ax[1,1].tick_params(axis='both', which='major', labelsize=34)

ax[1,1].legend(fontsize =34, loc="best")

for r1, r2, r3 in zip(bar1, bar2, bar3):
    h1 = r1.get_height()
    h2 = r2.get_height()
    h3 = r3.get_height()
    ax[1,1].text(r1.get_x() + r1.get_width() / 2., h1 + h1 / 2 , str(round(h1,2)) + " %", ha="center", va="center", color="white", fontsize = 35, fontweight="bold")
    ax[1,1].text(r2.get_x() + r2.get_width() / 2., h2/2 , str(round(h2,2)) + " %", ha="center", va="center", color="white", fontsize=35, fontweight="bold")
    ax[1,1].text(r3.get_x() + r3.get_width() / 2., h3 /2  , str(round(h3,2)) + " %", ha="center", va="center", color="white", fontsize=35, fontweight="bold")

# ax[1,1].title("2012 Olympics Top Scorers")




# ax[1,1].bar(ind, percGOP, width=0.8, label='GOP', color='r')
# ax[1,1].bar(ind, silvers, width=0.8, label='silvers', color='silver', bottom=bronzes)

# ax[1,1].xticks(ind, year)
# ax[1,1].ylabel("Medals")
# ax[1,1].xlabel("Countries")
# ax[1,1].legend(loc="upper right")
# ax[1,1].title("2012 Olympics Top Scorers")

plt.show()

#here we know that trump and biden are leading in all states but some they switch position to either first or second
# if (data["data"]["races"][3]["candidates"][0]["candidate_key"] == "bidenj"):
#     bidenVotePerc = round((100*float(data["data"]["races"][38]["candidates"][0]["votes"])/float(stateTotCountVote)),4)
#     trumpVotePerc = round((100*float(data["data"]["races"][38]["candidates"][1]["votes"])/float(stateTotCountVote)),4)
# else:
#     bidenVotePerc = round((100*float(data["data"]["races"][38]["candidates"][1]["votes"])/float(stateTotCountVote)),4)
#     trumpVotePerc = round((100*float(data["data"]["races"][38]["candidates"][0]["votes"])/float(stateTotCountVote)),4)

# reportingPerc = data["data"]["races"][3]["reporting_value"]
    
# print(reportingPerc)

# timeleng = len(data["data"]["races"][3]["timeseries"])
# bidenVotePercTime = data["data"]["races"][3]["timeseries"][3]["vote_shares"]["bidenj"]
# trumpVotePercTiem = data["data"]["races"][3]["timeseries"][3]["vote_shares"]["trumpd"]
# print(bidenVotePercTime,trumpVotePercTiem)

# x = []
# y = []

# index = count()
```
